## Implement Note Summarization

### Subtask:
Create a Python function that accepts study notes (text) as input and leverages an AI model to produce a concise summary of the key information.


**Reasoning**:
I will define the `summarize_notes` function as instructed, using the `gemma-3-1b-it` model which was successfully verified in the previous step, then create a prompt to summarize the input notes, and test the function with a sample text.



````python
import google.generativeai as genai

def summarize_notes(notes: str) -> str:
    """
    Summarizes study notes using an AI model.

    Args:
        notes (str): The study notes to be summarized.

    Returns:
        str: A concise summary of the notes.
    """
    # Load a generative AI model that supports content generation
    # Using 'gemma-3-1b-it' as it was successfully used for explanation.
    model = genai.GenerativeModel('gemma-3-1b-it')

    # Create a prompt for the AI model to summarize the notes concisely
    prompt = f"Please summarize the following study notes concisely, highlighting only the key information:\n\n{notes}"

    # Generate content from the model
    response = model.generate_content(prompt)

    # Extract and return the text summary
    return response.text

# Test the function with sample study notes
sample_notes = """
Quantum Entanglement:
- Two or more particles become linked, regardless of distance.
- Measuring one instantly determines the state of the other.
- Not used for faster-than-light communication; outcomes are random.
- Analogy: Two coins, one red, one blue, in separate boxes. Knowing one's color tells you the other's.
- Happens when particles are created together (e.g., decay).
- Foundation for quantum computing, cryptography, and teleportation (state transfer).
- Complex topic, still not fully understood.
"""

print("Summarizing study notes:")
summary = summarize_notes(sample_notes)
print(summary)
````

## Implement Quiz and Flashcard Generation

### Subtask:
Develop a Python function that takes study notes or a topic as input and utilizes an AI model to generate relevant quiz questions or flashcards.


**Reasoning**:
I will define the `generate_quiz_or_flashcards` function as instructed, using the `gemma-3-1b-it` model and constructing prompts based on the desired `output_format`. Then I will test the function with sample notes to generate both quiz questions and flashcards.



````python
import google.generativeai as genai

def generate_quiz_or_flashcards(input_text: str, output_format: str) -> str:
    """
    Generates quiz questions or flashcards from study notes or a topic using an AI model.

    Args:
        input_text (str): The study notes or topic to generate content from.
        output_format (str): 'quiz' to generate quiz questions, or 'flashcards' for flashcards.

    Returns:
        str: The generated quiz questions or flashcards in text format.
    """
    model = genai.GenerativeModel('gemma-3-1b-it')
    prompt = ""

    if output_format.lower() == 'quiz':
        prompt = f"""Generate 5 multiple-choice quiz questions with four options (A, B, C, D) and indicate the correct answer for each, based on the following notes. Provide the questions and answers clearly formatted:

{input_text}
"""
    elif output_format.lower() == 'flashcards':
        prompt = f"""Generate 5 flashcards from the following notes. Each flashcard should have a 'Front' (term/question) and a 'Back' (definition/answer). Format them clearly as 'Front: [text]\nBack: [text]\n\n':

{input_text}
"""
    else:
        return "Invalid output_format. Please choose 'quiz' or 'flashcards'."

    response = model.generate_content(prompt)
    return response.text

# Test the function with sample study notes for quiz questions
sample_notes = """
Quantum Entanglement:
- Two or more particles become linked, regardless of distance.
- Measuring one instantly determines the state of the other.
- Not used for faster-than-light communication; outcomes are random.
- Analogy: Two coins, one red, one blue, in separate boxes. Knowing one's color tells you the other's.
- Happens when particles are created together (e.g., decay).
- Foundation for quantum computing, cryptography, and teleportation (state transfer).
- Complex topic, still not fully understood.
"""

print("\n--- Generating Quiz Questions ---")
quiz_questions = generate_quiz_or_flashcards(sample_notes, 'quiz')
print(quiz_questions)

print("\n--- Generating Flashcards ---")
flashcards = generate_quiz_or_flashcards(sample_notes, 'flashcards')
print(flashcards)
````

## Create Interactive Interface

### Subtask:
Develop a basic interactive user interface within the notebook using ipywidgets to allow users to input text for explanation, summarization, or quiz generation, and display the AI-generated output.


**Reasoning**:
I will create the interactive user interface using ipywidgets, including input fields, buttons for each AI function, and an output area to display results. This involves importing necessary modules, defining widgets, creating callback functions to integrate with the AI functions, and arranging them in a user-friendly layout.



````python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Input widgets
text_input = widgets.Textarea(
    value='',
    placeholder='Enter your study notes or topic here...',
    description='Notes/Topic:',
    disabled=False,
    layout=widgets.Layout(width='auto', height='150px')
)

topic_input = widgets.Text(
    value='',
    placeholder='e.g., Quantum Entanglement',
    description='Topic for Explanation:',
    disabled=False,
    layout=widgets.Layout(width='auto')
)

# Output widget
output_area = widgets.Output()

# Function buttons
explain_button = widgets.Button(description='Explain Topic')
summarize_button = widgets.Button(description='Summarize Notes')
quiz_button = widgets.Button(description='Generate Quiz')
flashcard_button = widgets.Button(description='Generate Flashcards')

# Callback functions
def on_explain_button_clicked(b):
    with output_area:
        clear_output()
        topic = topic_input.value
        if topic:
            print(f"Explaining: {topic}\n")
            explanation = explain_topic(topic)
            print(explanation)
        else:
            print("Please enter a topic for explanation.")

def on_summarize_button_clicked(b):
    with output_area:
        clear_output()
        notes = text_input.value
        if notes:
            print("Summarizing study notes:\n")
            summary = summarize_notes(notes)
            print(summary)
        else:
            print("Please enter notes to summarize.")

def on_quiz_button_clicked(b):
    with output_area:
        clear_output()
        notes = text_input.value
        if notes:
            print("--- Generating Quiz Questions ---\n")
            quiz = generate_quiz_or_flashcards(notes, 'quiz')
            print(quiz)
        else:
            print("Please enter notes to generate quiz questions from.")

def on_flashcard_button_clicked(b):
    with output_area:
        clear_output()
        notes = text_input.value
        if notes:
            print("--- Generating Flashcards ---\n")
            flashcards = generate_quiz_or_flashcards(notes, 'flashcards')
            print(flashcards)
        else:
            print("Please enter notes to generate flashcards from.")

# Link buttons to callbacks
explain_button.on_click(on_explain_button_clicked)
summarize_button.on_click(on_summarize_button_clicked)
quiz_button.on_click(on_quiz_button_clicked)
flashcard_button.on_click(on_flashcard_button_clicked)

# Arrange widgets
input_box = widgets.VBox([
    topic_input,
    text_input
])

button_box = widgets.HBox([
    explain_button,
    summarize_button,
    quiz_button,
    flashcard_button
])

display(input_box, button_box, output_area)

print("Interactive AI Study Buddy Interface loaded. Please use the input fields and buttons above.")
````
